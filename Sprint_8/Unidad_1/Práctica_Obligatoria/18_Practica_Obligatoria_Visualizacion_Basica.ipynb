### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#Gráfico 1
top_populated_cities = df.nlargest(5, "population_total")

colors = ["blue", "gray", "olive", "brown", "pink"]

plt.figure(figsize=(8, 6))
bars = plt.bar(top_populated_cities["city"], top_populated_cities["population_total"], color=colors, alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,}', ha="center", va="bottom")

plt.ylabel("Número de personas (miles)")
plt.xlabel("Ciudades")

plt.show()
```

```python
#Gráfico 2
plt.figure(figsize=(8, 6))

plt.hist(df["area_total_km2"].dropna(), bins=50, color="steelblue", alpha=0.8, label="Hist Area")
plt.xlabel("Superficie (km²)")
plt.ylabel("Número de Ciudades")
plt.legend()

plt.show()
```

```python
#Gráfico 3
plt.figure(figsize=(8, 6))

plt.scatter(df["area_water_km2"], df["population_total"], color="green", alpha=0.6, edgecolor="black", label="Ciudades")

outliers = df[df["area_water_km2"] > 100]

for _, row in outliers.iterrows():
    plt.annotate("Outlier", 
                 xy=(row["area_water_km2"], row["population_total"]),  # Coordenadas del punto
                 xytext=(row["area_water_km2"] - 20, row["population_total"] + 500),  # Desplazamiento del texto
                 arrowprops=dict(facecolor="black", arrowstyle="->"),  # Estilo de la flecha
                 fontsize=10, color="black")

plt.title("Población por kilómetro cuadrado de agua", fontsize=14)
plt.xlabel("Área de agua en la ciudad (km²)")
plt.ylabel("Población total")
plt.legend()

plt.show()
```

```python
#Gráfico 4
plt.figure(figsize=(8, 6))

plt.boxplot(df["elevation_m"].dropna(), patch_artist=True,boxprops=dict(facecolor="red", color="black"), vert=False, medianprops=dict(color="yellow", linewidth=2))

plt.title("Diagrama de caja de la Altitud de las Ciudades de California", fontsize=14)
plt.xlabel("Ciudades")
plt.ylabel("Altitud (m)")

# Mostrar el gráfico
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Gráfico 1
top_populated_cities = df.nlargest(5, "population_total")
colors = ["blue", "gray", "olive", "brown", "pink"]
bars = axes[0, 0].bar(top_populated_cities["city"], top_populated_cities["population_total"], color=colors, alpha=0.7)
for bar in bars:
    height = bar.get_height()
    axes[0, 0].text(bar.get_x() + bar.get_width() / 2, height, f'{height:,}', ha="center", va="bottom")
axes[0, 0].set_title("Ciudades más pobladas de California")
axes[0, 0].set_ylabel("Número de personas (miles)")
axes[0, 0].set_xlabel("Ciudades")

# Gráfico 2
axes[0, 1].hist(df["area_total_km2"].dropna(), bins=50, color="steelblue", alpha=0.8, label="Hist Área")
axes[0, 1].set_xlabel("Superficie (km²)")
axes[0, 1].set_ylabel("Número de Ciudades")
axes[0, 1].legend()

# Gráfico 3
axes[1, 0].scatter(df["area_water_km2"], df["population_total"], 
                   color="green", alpha=0.6, edgecolor="black", label="Ciudades")
axes[1, 0].set_title("Población por kilómetro cuadrado de agua", fontsize=14)
axes[1, 0].set_xlabel("Área de agua en la ciudad (km²)")
axes[1, 0].set_ylabel("Población total")
axes[1, 0].legend()

# Gráfico 4
axes[1, 1].boxplot(df["elevation_m"].dropna(), patch_artist=True,
                   boxprops=dict(facecolor="red", color="black"),vert=False, medianprops=dict(color="yellow", linewidth=2))
axes[1, 1].set_title("Diagrama de caja de la Altitud de las Ciudades de California", fontsize=14)
axes[1, 1].set_xlabel("Ciudades")
axes[1, 1].set_ylabel("Altitud (m)")

plt.tight_layout()

plt.savefig("california_cities_analysis.png")

# Mostrar la figura
plt.show()
```
