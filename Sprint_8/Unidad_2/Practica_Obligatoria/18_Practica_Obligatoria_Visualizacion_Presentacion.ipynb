### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def pintar_dispersin_avanzada(df, col_x, col_y, col_cat=None, punto_tam=50, escala=1, mostrar_leyenda=True):

    if isinstance(punto_tam, str):
        if punto_tam in df.columns:
            tamaos = df[punto_tam] * escala
        else:
            raise ValueError(f"La columna '{punto_tam}' no existe en el DataFrame.")
    else:  
        tamaos = punto_tam

    plt.figure(figsize=(12, 8))

    # Graficar con o sin columna categrica
    if col_cat:  # Con columna categrica
        scatter = sns.scatterplot(
            data=df,
            x=col_x,
            y=col_y,
            hue=col_cat,
            size=tamaos if isinstance(punto_tam, str) else None,
            sizes=(20, 200) if isinstance(punto_tam, str) else None,
            alpha=0.8,
            palette="viridis"
        )
        if mostrar_leyenda:
            scatter.legend(title=col_cat, loc='best')
        else:
            # Si no se quiere una leyenda estndar, se muestra un colorbar
            norm = plt.Normalize(df[punto_tam].min() * escala, df[punto_tam].max() * escala)
            sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
            sm.set_array([])
            plt.colorbar(sm, label=punto_tam)
    else:  # Sin columna categrica
        plt.scatter(
            df[col_x],
            df[col_y],
            c='blue',
            s=tamaos,
            alpha=0.8
        )

    plt.title(f'Diagrama de dispersin: {col_x} vs {col_y}', fontsize=16)
    plt.xlabel(f'{col_x}', fontsize=12)
    plt.ylabel(f'{col_y}', fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```

```python
titanic_df = pd.read_csv("/Users/alvarosanchez/ONLINE_DS_THEBRIDGE_ALVAROSMMS-1/ONLINE_DS_THEBRIDGE_ALVAROSMMS/Sprint_8/Unidad_2/Practica_Obligatoria/data/titanic.csv")
```

```python
#1- Tasa de mortalidad alta.
tasa_mortalidad = titanic_df['survived'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=["No sobrevivi","Sobrevivi"], y=tasa_mortalidad.values, palette=sns.color_palette("coolwarm",n_colors=2))
plt.ylabel('Porcentaje')
plt.title('Tasa de Mortalidad en el Titanic')
plt.tight_layout()

plt.savefig('tasa_mortalidad', dpi=300, bbox_inches='tight')
```

```python
#2-Los datos apoyan "las mujeres y los nios primero" pero...
titanic_df['child'] = titanic_df['age'].apply(lambda x: 'Child' if x < 16 else 'Adult')
supervivencia_sexo = titanic_df.groupby(['sex', 'child'])['survived'].mean() * 1000

plt.figure(figsize=(10, 6))
sns.barplot(data=titanic_df, x='sex', hue='child', y='survived', errorbar=None, palette="coolwarm")
plt.xlabel('Sexo')
plt.ylabel('Tasa de Supervivencia (%)')
plt.title('Tasa de Supervivencia: Mujeres y Nios Primero')
plt.tight_layout()

plt.savefig('mujeres_y_nios', dpi=300, bbox_inches='tight')
```

```python
#3
supervivencia_clase = titanic_df.groupby('pclass')['survived'].mean() * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=supervivencia_clase.index, y=supervivencia_clase.values, palette="coolwarm")
plt.xticks([0, 1, 2], ['Primera Clase', 'Segunda Clase', 'Tercera Clase'])
plt.xlabel('Clase')
plt.ylabel('Tasa de Supervivencia (%)')
plt.title('Supervivencia por Clase')
plt.tight_layout()

plt.savefig('influencia_clase', dpi=300, bbox_inches='tight')
```

```python
# 4. Tasa de Supervivencia de Mujeres y Nios por Clase
titanic_df['child'] = titanic_df['age'].apply(lambda x: 'Child' if x < 16 else 'Adult')

mujeres_ninos = titanic_df[(titanic_df['sex'] == 'female') | (titanic_df['child'] == 'Child')]

supervivencia_mujeres_ninos = mujeres_ninos.groupby(['pclass', 'child', 'sex'])['survived'].mean() * 100

plt.figure(figsize=(12, 8))
sns.barplot(
    data=mujeres_ninos, 
    x='pclass', 
    y='survived', 
    hue='child', 
    palette="coolwarm", 
    errorbar=None
)
plt.ylabel('Tasa de Supervivencia (%)')
plt.xlabel('Clase')
plt.title('Supervivencia entre Mujeres y Nios por Clase')
plt.tight_layout()

plt.savefig('supervivencia_por_clase', dpi=700, bbox_inches='tight')

plt.show()
```

```python
from matplotlib.colors import ListedColormap

#5
#Proporcin de clases por puerto de embarque
proporcion_clases_puerto = titanic_df.groupby(['embark_town', 'pclass']).size().unstack().apply(lambda x: x / x.sum(), axis=1) * 100

#Tasa de supervivencia por puerto y clase
tasa_supervivencia_puerto_clase = titanic_df.groupby(['embark_town', 'pclass'])['survived'].mean().unstack() * 100

fig, axes = plt.subplots(2, 1, figsize=(12, 12), sharex=True)

pastel_cmap = ListedColormap(['#add8e6', '#ffb6c1', '#c3e5d9']) 

proporcion_clases_puerto.plot(kind='bar', stacked=True, colormap=pastel_cmap, ax=axes[0])
axes[0].set_title('Distribucin de Clases por Puerto de Embarque')
axes[0].set_ylabel('Proporcin (%)')
axes[0].legend(title='Clase', labels=['Primera Clase', 'Segunda Clase', 'Tercera Clase'])

tasa_supervivencia_puerto_clase.plot(kind='bar', colormap=pastel_cmap, ax=axes[1])
axes[1].set_title('Tasa de Supervivencia por Puerto y Clase')
axes[1].set_ylabel('Tasa de Supervivencia (%)')
axes[1].legend(title='Clase', labels=['Primera Clase', 'Segunda Clase', 'Tercera Clase'])

plt.xlabel('Puerto de Embarque')
plt.xticks(rotation=0)  
plt.tight_layout()


plt.savefig('supervivencia_por_clase', dpi=700, bbox_inches='tight')

plt.show()

```
